```python
import pandas as pd
df=pd.read_csv("https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Class",data=df)
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=5)
k.fit(x_train,y_train)
k1=k.score(x_test,y_test)
k1
```

```python
# Balance the dataset using Random Over sampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(x_train, y_train)

#class wise disctributions in train data split after random over sampling
print("class wise disctributions in train data split after random over sampling")
print(y_resampled.value_counts())


# Implement KNN model for classifying data balanced using Random Over sampling
knn_ros = KNeighborsClassifier(n_neighbors=5)
knn_ros.fit(X_resampled, y_resampled)
ros_score = knn_ros.score(x_test, y_test)
print('KNN accuracy on Random Over sampled data:', ros_score)
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler(random_state=42)
x1_re,y1_re=rus.fit_resample(x_train,y_train)
print(y1_re.value_counts())
k=KNeighborsClassifier(n_neighbors=5)
k.fit(x1_re,y1_re)
print(k.score(x_test,y_test))
```

```python
from imblearn.over_sampling import SMOTE
s=SMOTE(random_state=42)
x2_re,y2_re=s.fit_resample(x_train,y_train)
print(y2_re.value_counts())
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df.head
```

```python
x2=df.iloc[:,0]
y2=df.iloc[:,1]
x2
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  k=KMeans(n_clusters=i,init="k-means++",max_iter=300)
  k.fit(x)
  wcss.append(k.inertia_)
print(wcss) 
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,max_iter=20,init="k-means++")
kmeans.fit(x)
labels=kmeans.predict(x) 
centroids=kmeans.cluster_centers_ 
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=labels)
plt.scatter(centroids[:,0],centroids[:,1],color="red",s=200,linewidth=3,marker="*")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.5,min_samples=5)
db.fit(x)
labels=db.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print(n_clusters_)
print(n_noise_)
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=labels)
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,train_size=0.8)
pca=PCA(n_components=2)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
x_test_pca[:,0]
```

```python
import pandas as pd
import numpy as np

# create dataframes from the numpy arrays
x_train_df = pd.DataFrame(x_train_pca)
y_train_df = pd.DataFrame(y_train)
x_test_df = pd.DataFrame(x_test_pca)
y_test_df = pd.DataFrame(y_test)

# add the target variable column to the train and test dataframes
x_train_df['target'] = y_train_df
x_test_df['target'] = y_test_df

# concatenate the train and test dataframes vertically
merged_df = pd.concat([x_train_df, x_test_df], axis=0)

# reset the index of the merged dataframe
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
merged_df=merged_df.dropna()
```

```python
merged_df.isna().sum()
```

```python
merged=merged_df.iloc[:,:-1]
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++",max_iter=10)
kmeans.fit(merged)
labels=kmeans.predict(merged)
centroids=kmeans.cluster_centers_
plt.scatter(merged.iloc[:,0],merged.iloc[:,1],c=labels)
plt.scatter(centroids[:,0],centroids[:,1],marker="*",linewidth=3,color="red")
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
dendro=shc.dendrogram(shc.linkage(merged,method="ward")) 
plt.show() 
```

```python
from sklearn.cluster import AgglomerativeClustering
ahc=AgglomerativeClustering(affinity="euclidean",n_clusters=2,linkage="ward")
labels=ahc.fit_predict(merged)
plt.scatter(merged.iloc[:,0],merged.iloc[:,1],c=labels) 
plt.show() 
```
